```python
class Kiosk:
    def __init__(self):
        self.menu = ['아메리카노','카페라떼','브류잉커피','초코라떼','말차라떼','녹차']
        self.price = [5000, 6000, 7000, 6000, 6500, 5000]
        self.order_menu = []
        self.order_price = []

    def menu_print(self):
        for i in range(len(self.menu)):
            print(i+1, self.menu[i], ' : ', self.price[i], '원')

    def menu_select(self):
        print()
        n = 0
        while not (1 <= n <= len(self.menu)):
            try:
                n = int(input("음료주문은 숫자로 해주세요: "))
                if 1 <= n <= len(self.menu):
                    self.price_sum = self.price[n-1]
                else:
                    print("주문이 잘못되었습니다.")
                    n = 0  
            except:
                print("주문이 잘못되었습니다.")
                n = 0

        t = 0
        while t != 1 and t != 2:  
            try:
                t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요: "))
                if t == 1:         
                    self.temp = "HOT"
                elif t == 2:
                    self.temp = "ICE"
                else:
                    print("1과 2 중 하나를 입력하세요.\n")
                    t = 0
            except:
                print("잘못된 입력입니다.")
                t = 0

        print('주문 음료', self.temp, self.menu[n-1], ' : ', self.price_sum, '원')
        self.order_menu.append(f"{self.temp} {self.menu[n-1]}")
        self.order_price.append(self.price_sum)

    def add_menu(self):  
        print()
        self.menu_select()
        answer = 3
        while answer not in [0,1]:
            try:
                answer = int(input("결제는 0번 추가주문은 1번을 눌러주세요."))
                if answer == 0:
                    print("결제를 하겠습니다.")
                    self.pay()
                    
                elif answer == 1:
                    print("추가주문을 하겠습니다.")
                    self.menu_select()
                    answer = 3
                else:
                    print("잘못된 입력입니다.")
                    answer = 3
            except :
                print("입력이 잘못 됬습니다.")  
                answer = 3

        

    def pay(self):
        for om in range(len(self.order_menu)):
            print(om+1,self.order_menu[om],self.order_price[om],"원")
        print("결제 총액:", sum(self.order_price),"원")
        p=0
        while p not in [1,2]:
            try:
                p = int(input("현금은1번 카드는 2번을 눌러주세요."))
                if p == 1:
                    print('직원을 호출하겠습니다.' )
                    self.checkout = '현금'         
              
                elif p == 2:
                    print('IC칩 방향에 맞게 카드를 꽂아주세요.')
                    self.checkout = '카드'    
           
                else:
                    print("입력이 잘못됬습니다.")
                    p=0
            except:
                print("입력을 잘못 했습니다.")
                p=0
        self.table()
    
    def table(self):
        print('⟝' + '-' * 30 + '⟞')      
        
        for om in range(len(self.order_menu)):
            print("  ",om+1,self.order_menu[om],self.order_price[om])
            
        print('='*33)
        print("결제방식:",self.checkout)
        print('합계 금액 :', sum(self.order_price))
    
        print('⟝' + '-' * 30 + '⟞')

```

```python
kio= Kiosk()
kio.menu_print()
kio.add_menu()
```
